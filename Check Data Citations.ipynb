```python
import pandas as pd
import numpy as np
from ast import literal_eval
from tqdm import tqdm

pd.options.display.max_colwidth = 1000
```

```python
%%time

# Read information for the new records
records = pd.read_csv('new-citations-data.tsv', sep='\t', header=None)
records.columns = ['recID', 'oldRef', 'newRef', 'metaData']

print('Total Records: {}'.format(len(records)))
```

```python
data_citations = records[records.newRef != 0]

# Filter out other records not captured by legacy
new_captures = data_citations[data_citations.oldRef != data_citations.newRef]
```

```python
print('Total {} data citations captured by Labs.'.format(len(data_citations)))
print('Total {} new data citations captured by Labs which were missed by legacy.'.format(len(new_captures)))
```
