```python
import numpy as np
import pandas as pd
from collections import defaultdict
import ujson
import requests

pd.options.display.max_colwidth = 1000
```

```python
# Read the records
records = pd.read_csv('new-citations.tsv', sep='\t', header=None)
```

```python
# Get the records which have a different result for the legacy alogorithm and the new algorithm
not_equal = records[1] != records[2]
wrongs = records[not_equal]

# Filter out records representative of Errata i.e. Old Algorithm result = 0 and New Algorithm result != 0
not_errata = wrongs[2] != 0
errata_wrongs = wrongs[not_errata]
old_zeros = errata_wrongs[1] == 0
errata_wrongs2 = errata_wrongs[old_zeros]
```

### Verify the errata references

```python
# Function to query the API if a reference is an erratum
def check_erratum(ref):
    '''
        Takes a reference ID as input and returns whether it is erratum or not according 
        to the INSPIRE Labs API recrod for that reference.
        
        Input: (int) ref_id
        Output: (Bool) Indiciating True or False for Erratum or not respectively
    '''
    
    url = 'http://labs.inspirehep.net/api/literature/'+str(ref)+'/db' 
    json_data = requests.get(url).json()
    try:
        if json_data['metadata']['publication_info'][1]['material'] == 'erratum':
            return True
    except:
        return False
    
# Function to query the API if a reference is JHEP
def check_jhep(ref):
    '''
        Takes a reference ID as input and returns whether it is from JHEP or not according 
        to the INSPIRE Labs API recrod for that reference.
        
        Input: (int) ref_id
        Output: (Bool) Indiciating True or False for JHEP or not respectively
    '''
    
    url = 'http://labs.inspirehep.net/api/literature/'+str(ref)+'/db' 
    json_data = requests.get(url).json()
    try:
        if json_data['metadata']['publication_info'][0]['journal_title'] == 'JHEP':
            return True
    except:
        return False
    
# Function to query the API if a reference is JCAP
def check_jcap(ref):
    '''
        Takes a reference ID as input and returns whether it is from JCAP or not according 
        to the INSPIRE Labs API recrod for that reference.
        
        Input: (int) ref_id
        Output: (Bool) Indiciating True or False for JCAP or not respectively
    '''
    
    url = 'http://labs.inspirehep.net/api/literature/'+str(ref)+'/db' 
    json_data = requests.get(url).json()
    try:
        if json_data['metadata']['publication_info'][0]['journal_title'] == 'JCAP':
            return True
    except:
        return False
```

```python
%%time

# Iterate over all the wrongs and identify the errata references using the INSPIRE LABS API
errt_count = 0
jhep_count = 0
jcap_count = 0
other_fp = 0

errt_dict = {}
jhep_dict = {}
jcap_dict = {}

for rec, refs in dup_dict.items():
    
    for ref in refs:

        rIDNew = ref.split('/')[-2]
        # Check the reference from the API if they are errata
        errt = check_erratum(rIDNew)
        if errt:
            print('Record: {} - Result: {} is ERRATUM'.format(rec, rIDNew))
            errt_dict.setdefault(rec, []).append('http://labs.inspirehep.net/api/literature/'+str(rIDNew)+'/db')
            errt_count += 1
        else:
            # Check if JHEP
            jhep = check_jhep(rIDNew)
            # Check if JCAP
            jcap = check_jcap(rIDNew)
            if jhep:
                print('JHEP Detected - Record: {} - Result: {}'.format(rec, rIDNew))
                jhep_dict.setdefault(rec, []).append('http://labs.inspirehep.net/api/literature/'+str(rIDNew)+'/db')
                jhep_count += 1
            elif jcap:
                print('jcap detected - Record: {} - Result: {}'.format(rec, rIDNew))
                jcap_dict.setdefault(rec, []).append('http://labs.inspirehep.net/api/literature/'+str(rIDNew)+'/db')
                jcap_count += 1
            else:
                print('Record: {} - Result: {} is NOT ERRT, JHEP, OR JHEP'.format(rec, rIDNew))
                other_fp += 1
            
                
print('Errata Count: {}, Errt Dict Count: {}'.format(errt_count, len(errt_dict)))
print('JHEP Count: {}, JHEP Dict Count: {}'.format(jhep_count, len(jhep_dict)))
print('JCAP Count: {}, JCAP Dict Count: {}'.format(jcap_count, len(jcap_dict)))
print('Other False Positives Count: {}'.format(other_fp))
print()
```
